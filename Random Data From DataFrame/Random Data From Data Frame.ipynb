```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'a': np.random.choice(np.random.randint(1000, size=1000), size=100),
    'b': np.random.choice(np.random.randint(1000, size=1000), size=100),
    'c': np.random.choice(np.random.randint(1000, size=1000), size=100)
}
df = pd.DataFrame(data)
df.head()
```

```python
# random = np.random.choice([True, False], size=100)
# df[random]
# d = df[random]
# d.head()
# d.loc[:50]

df.loc[np.random.choice(df.index, size=50)]
```

```python
plt.style.use('fivethirtyeight')
plt.hist(df.loc[np.random.choice(df.index, size=50)], label=['A', 'B', 'C'])
plt.legend(loc='upper left')
plt.title("just my Random Number")
plt.xlabel('Number')
plt.ylabel('Leve')
plt.grid(True)
plt.show()
```
